```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('./csv_new/open.csv')
df = pd.DataFrame(data)
```

```python
df['tag'] = ''

def add_equipment (contain_str, add_str):
    for index, row in df.iterrows():
        if (row['tag'] == '') and (contain_str in row['workoutDetails'].lower()) and (add_str not in row['tag']):
            df.at[index, 'tag'] = row['tag'] + add_str
        elif (row['tag'] != '') and (contain_str in row['workoutDetails'].lower()) and (add_str not in row['tag']):
            df.at[index, 'tag'] = row['tag'] + ', '+ add_str

add_equipment('toes', 'PULL UP BAR')
add_equipment('dumbbell', 'DUMBBELL')
add_equipment('jerk', 'BARBELL')
add_equipment('row', 'ROWER')
add_equipment('clean', 'BARBELL')
add_equipment('double-unders', 'SKIPPING ROPE')
add_equipment('deadlift', 'BARBELL')
add_equipment('overhead squats', 'BARBELL')
add_equipment('ring', 'GYM RINGS')
add_equipment('muscle-up', 'PULL UP BAR')
add_equipment('thrusters', 'BARBELL')
add_equipment('pull up', 'PULL UP BAR')
add_equipment('box jump', 'BOX')
add_equipment('snatch', 'BARBELL')
add_equipment('wall ball', 'MED BALL')
add_equipment('dus', 'SKIPPING ROPE')
add_equipment('ttb', 'PULL UP BAR')
add_equipment('muscle up', 'PULL UP BAR')
add_equipment('shoulder-to-overhead', 'BARBELL')
add_equipment('bar-facing', 'BARBELL')
add_equipment('bar facing', 'BARBELL')
df.at[5, 'tag'] = 'DUMBBELL, BOX'
df.at[6, 'tag'] = 'PULL UP BAR, DUMBBELL'
```
